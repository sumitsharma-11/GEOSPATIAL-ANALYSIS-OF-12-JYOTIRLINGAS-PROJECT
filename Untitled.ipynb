```python
from geopy.geocoders import Nominatim
import pandas as pd

geolocator = Nominatim(user_agent="shiva_project")
temples = [
    "Somnath Temple, Gujarat", "Mallikarjuna Temple, Srisailam", 
    "Mahakaleshwar, Ujjain", "Omkareshwar, Madhya Pradesh", 
    "Kedarnath Temple, Uttarakhand", "Bhimashankar Temple, Maharashtra", 
    "Kashi Vishwanath, Varanasi", "Trimbakeshwar, Nashik", 
    "Vaidyanath Temple, Deoghar" , "Nageshwar Jyotirlinga, Gujarat", 
    "Ramanathaswamy Temple, Rameshwaram", "Grishneshwar Jyotirlinga Temple, Maharashtra"
]

manual_coords = {
    "Vaidyanath Temple, Deoghar": (24.4931, 86.7000),
    "Nageshwar Jyotirlinga, Gujarat": (22.3359, 69.0869),
    "Grishneshwar Jyotirlinga Temple, Maharashtra": (20.0249, 75.1698)
}

data = []
for t in temples:
    location = geolocator.geocode(t)
    if location:
        data.append({"Temple": t, "Lat": location.latitude, "Lon": location.longitude})
    elif t in manual_coords:
        # Use our manual data if the geocoder fails
        lat, lon = manual_coords[t]
        data.append({"Temple": t, "Lat": lat, "Lon": lon})
    else:
        print(f"Still could not find: {t}")

df = pd.DataFrame(data)
df.to_csv("jyotirlinga_coords.csv", index=True)
```

```python
# Logical check for Lat/Lon boundaries
invalid_coords = df[(df['Lat'] < -90) | (df['Lat'] > 90) | 
                    (df['Lon'] < -180) | (df['Lon'] > 180)]

if not invalid_coords.empty:
    print("Action Required: Invalid coordinates detected for:", invalid_coords['Temple_Name'].tolist())
else:
    print("All coordinates are within valid Earth boundaries.")
```
